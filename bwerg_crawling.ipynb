```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

language_dic = {'deutsch':'https://www.bverwg.de/',
                'english':'https://www.bverwg.de/en',
                'france':'https://www.bverwg.de/fr'} 

crawling_data = {}

for language, url in language_dic.items():
  urls = url
  driver.get(urls)
  element = driver.find_element(By.XPATH,"/html")

  element_text = element.text

  text_list = element_text.split('\n')

  # crawling_data[language] = text_list


  file_name = f"./resource/{language}_crawled_data.txt"
  with open(file_name, 'w',encoding='utf-8') as file:
    for line in text_list:
      file.write(line+'\n')


  # print(element_text)
driver.quit()
```
